# OpenAI TTS API 사용 방법 알아보기

## 소개
Audio API는 TTS(텍스트 음성 변환) 모델을 기반으로 음성 엔드포인트를 제공합니다. 6개의 내장 음성이 제공되며 다음과 같은 용도로 사용할 수 있습니다.

- 작성된 블로그 게시물 설명
- 여러 언어로 음성 오디오 생성
- 스트리밍을 사용하여 실시간 오디오 출력 제공


```python
from openai import OpenAI
```

```python
client = OpenAI()
```

```python
!mkdir outputs
```

```python
response = client.audio.speech.create(
  model="tts-1",
  voice="nova", # alloy, echo, fable, , nova, and shimmer
  input="안녕하세요 강사 리암입니다."
)

speech_file_path = "outputs/sample_speech.mp3"
response.stream_to_file(speech_file_path)
```

```python
from IPython.display import Audio

# Audio 객체를 생성하고 출력합니다.
audio = Audio(speech_file_path, autoplay=False)
audio
```
